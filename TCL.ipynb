# Théorème Central Limite (TCL)

Le **théorème central limite** (TCL) est un résultat fondamental en statistique et en théorie des probabilités. Il énonce que, sous certaines conditions, la somme (ou la moyenne) d'un grand nombre de variables aléatoires indépendantes et identiquement distribuées (i.i.d.) suit approximativement une distribution **normale** (gaussienne), quelle que soit la distribution d'origine.

## Énoncé formel

Soient $( X_1, X_2, ...X_n)$ des variables aléatoires indépendantes et identiquement distribuées, d'espérance $ \mu $  et de variance $ \sigma^{2} $ fnie. On définit la somme  $S_n = X_1 + X_2 + \dots + X_n $ et la moyenne  $\bar{X}_n = \frac{S_n}{n}$.

Alors, lorsque $ n $  tend vers l'infini, la distribution de la moyenne $ \bar{X}_n $ converge vers une distribution normale d'espérance $ \mu $ et de variance $ \frac{\sigma^2}{n} $. Mathématiquement, cela s'écrit 

$$
\bar{X}_n \xrightarrow[n \to \infty]{\text{loi}} \mathcal{N}\left(\mu, \frac{\sigma^2}{n}\right)
$$

ou, de manière équivalente :

$$
\frac{\bar{X}_n - \mu}{\frac{\sigma}{\sqrt{n}}} \xrightarrow[n \to \infty]{\text{loi}} \mathcal{N}(0, 1)
$$

où $ \mathcal{N}(\mu, \sigma^2) $ désigne une distribution normale d'espérance $ \mu $ et de variance $ \sigma^2 $.

## Interprétation

Le TCL explique pourquoi de nombreux phénomènes naturels suivent une distribution normale, même lorsque les variables sous-jacentes ne sont pas normales. Par exemple :
- La moyenne des résultats d'un grand nombre de lancers de dés.
- La somme des erreurs de mesure dans une expérience scientifique.
- La distribution des moyennes d'échantillons dans les enquêtes statistiques.

## Conditions d'application

Pour que le TCL s'applique, les conditions suivantes doivent être respectées :
1. **Indépendance** : Les variables $( X_1, X_2, \dots, X_n )$ doivent être indépendantes.
2. **Identiquement distribuées** : Les variables doivent avoir la même distribution.
3. **Variance finie** : La variance $ \sigma^2 $ doit être finie (non infinie).

## Exemple

Supposons que nous lancions un dé équilibré $ (n = 100) $ fois et que nous calculions la moyenne des résultats. D'après le TCL, la distribution de cette moyenne sera approximativement normale, avec :
- Espérance $( \mu = 3.5 )$ (la moyenne d'un dé équilibré).
- Variance $( \frac{\sigma^2}{n} = \frac{2.92}{100} = 0.0292 )$ (où $( \sigma^2 = 2.92 )$ est la variance d'un dé équilibré).



# Théorème Central Limite (TCL) : Visualisation 

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import random

n=10000
l = np.zeros(n)
M = np.zeros(n)
for j in range(0,n):
    for i in range(0,n):
        l[i] = random.randint(0,200)
    s=l.sum()
    M[j]=s/n
print(M)

plt.figure(figsize=(15,8))
sns.histplot(M,kde=True,color="orange")
plt.title("DISTRIBUTION NORMALE")
plt.show()
```
